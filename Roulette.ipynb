```python
import random

money = 100
def roulette(bet, guess):
    print('-------')
    print('Welcome to the game of Roulette')
    print('Here is $100, make the most of it!')
    print('-------')
    print('------')
    print('Lets spin the wheel!')
    print('------')
    if bet <= 0:
        print('You need to bet something!')
        return 0
    result = random.randint(0,37)
    if result == 37:
        print('00 bad news! You lose')
        return -bet
    elif result == guess:
        print ('Congratulations! You have won ' + str(bet * 35) + ' dollars!')
        return bet * 35
    else:
        print('you guessed ' + str(guess) + ' and it was ' + str(result) + ' better luck next time!')
        return -bet

money += roulette(100,25)

print(money)
```
